```python
import yfinance as yf
import pandas as pd
from statsmodels.tsa.arima.model import ARIMA

# Fetch historical stock data for Baidu (BIDU)
baidu_data = yf.download('BIDU', start='2015-01-01', end='2023-12-31')['Adj Close']

# Fetch historical data for USD/CNY exchange rate
usdcny_data = yf.download('CNY=X', start='2015-01-01', end='2023-12-31')['Adj Close']

# Combine stock and currency data into 1 datafrake
data = pd.concat([baidu_data, usdcny_data], axis=1)
data.columns = ['BIDU', 'USD_CNY']

data = data.dropna() # missing vals

data = data.resample('M').mean() # M for monthly freq

# Fit ARIMA model for Baidu stock prices
model_bidu = ARIMA(data['BIDU'], order=(5, 1, 0))
results_bidu = model_bidu.fit()

# Forecast
forecast_bidu = results_bidu.forecast(steps=10)  # Forecast 10 steps ahead
print("Forecasted BIDU stock prices:")
print(forecast_bidu)
```

feature engineering

```python
# Create lag features and rolling stats
for i in range(1, 6):  #  up to 5 days
    data[f'BIDU_lag_{i}'] = data['BIDU'].shift(i)
    data[f'USD_CNY_lag_{i}'] = data['USD_CNY'].shift(i)

rolling_window = 5
data['BIDU_mean_rolling'] = data['BIDU'].rolling(window=rolling_window).mean()
data['BIDU_std_rolling'] = data['BIDU'].rolling(window=rolling_window).std()
data['USD_CNY_mean_rolling'] = data['USD_CNY'].rolling(window=rolling_window).mean()
data['USD_CNY_std_rolling'] = data['USD_CNY'].rolling(window=rolling_window).std()

print(data.head(10))  # check first 10 new features/rows
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

data.dropna(inplace=True)

# Features (lagged USD/CNY exchange rates) and target (Baidu stock prices)
features = [col for col in data.columns if 'USD_CNY_lag_' in col]
target = 'BIDU'

X_train, X_test, y_train, y_test = train_test_split(data[features], data[target], test_size=0.2, random_state=42)

rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)

predictions = rf_model.predict(X_test)

# MSE evaluation
mse = mean_squared_error(y_test, predictions)
print(f"Mean Squared Error: {mse}")
```

```python
# Optimize model to lower MSE
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor

param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 5, 10, 15]
}

rf_model = RandomForestRegressor(random_state=42)

grid_search = GridSearchCV(rf_model, param_grid, cv=5, scoring='neg_mean_squared_error')
grid_search.fit(X_train, y_train)

best_params = grid_search.best_params_
best_rf_model = RandomForestRegressor(**best_params, random_state=42)
best_rf_model.fit(X_train, y_train)

predictions = best_rf_model.predict(X_test)

mse = mean_squared_error(y_test, predictions)
print(f"Mean Squared Error: {mse}")
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Remove NaN values
data_clean = data.dropna()
data_clean = data_clean[(np.abs(data_clean['BIDU_percent_change'] - data_clean['BIDU_percent_change'].mean()) < 3 * data_clean['BIDU_percent_change'].std()) &
                        (np.abs(data_clean['USD_CNY_percent_change'] - data_clean['USD_CNY_percent_change'].mean()) < 3 * data_clean['USD_CNY_percent_change'].std())]

plt.figure(figsize=(8, 6))
plt.scatter(data_clean['USD_CNY_percent_change'], data_clean['BIDU_percent_change'], color='blue', alpha=0.5)
plt.xlabel('USD/CNY Exchange Rate Percent Change')
plt.ylabel('Baidu Stock Price Percent Change')
plt.title('Relationship between USD/CNY Exchange and Baidu Stock Price Percent Change')
plt.grid(True)

# line of best fit
slope, intercept = np.polyfit(data_clean['USD_CNY_percent_change'], data_clean['BIDU_percent_change'], 1)
plt.ylim(data['BIDU_percent_change'].min() - 20, data['BIDU_percent_change'].max() + 20)
plt.plot(data_clean['USD_CNY_percent_change'], slope * data_clean['USD_CNY_percent_change'] + intercept, color='red', label='Line of Best Fit')

plt.legend()
plt.show()


predicted_values = slope * data_clean['USD_CNY_percent_change'] + intercept

# Compute the Mean Squared Error (MSE)
actual_values = data_clean['BIDU_percent_change']  # Actual values
mse = ((predicted_values - actual_values) ** 2).mean()
rmse = np.sqrt(mse)
print("Root Mean Squared Error (RMSE):", rmse)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import linregress

# Assuming 'data' is the DataFrame containing 'BIDU_percent_change' and 'USD_CNY_percent_change' columns

# Fit a linear regression model to get the slope and intercept
slope, intercept, _, _, _ = linregress(data['USD_CNY_percent_change'], data['BIDU_percent_change'])

# Calculate residuals (vertical distances) of each data point from the line of best fit
residuals = data['BIDU_percent_change'] - (slope * data['USD_CNY_percent_change'] + intercept)

# Calculate absolute residuals
absolute_residuals = np.abs(residuals)

# Find the indices of points with the largest residuals to remove
num_points_to_remove = 5  # Number of points to remove (change as needed)
indices_to_remove = absolute_residuals.nlargest(num_points_to_remove).index

# Remove the furthest points from the line of best fit
data_clean = data.drop(indices_to_remove)

# Calculate the correlation coefficient between the cleaned data
correlation_coefficient = data_clean['BIDU_percent_change'].corr(data_clean['USD_CNY_percent_change'])

# Plot scatter plot of cleaned data
plt.figure(figsize=(8, 6))
plt.scatter(data_clean['USD_CNY_percent_change'], data_clean['BIDU_percent_change'], color='blue', alpha=0.5)
plt.xlabel('USD/CNY Exchange Rate Percent Change')
plt.ylabel('Baidu Stock Price Percent Change')
plt.title('Relationship between USD/CNY Exchange and Baidu Stock Price Percent Change (Outliers Removed)')
plt.grid(True)
plt.show()

print("Correlation coefficient after removing furthest points from the line of best fit:", correlation_coefficient)
```
